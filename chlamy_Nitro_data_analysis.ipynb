```python
import pybedtools
import sys
import multiprocessing
import cStringIO
import tempfile
import matplotlib.pyplot as plt
```

```python
#a function to pass only features of a particular featuretype
#this is similar to grep
# if inverse_flag is set to True, this is similar to grep -v
def featuretype_filter(feature,featuretypes,inverse_flag=False):
    for featuretype in featuretypes:
        if feature[2] == featuretype:
            #match found
            if inverse_flag is True:
                return False
            else:
                return True
            
    #no match found
    if inverse_flag is True:
        return True
    else:
        return False


#function to subset a feature file selecting the features 
def subset_featuretypes(bedToolsFile,featuretypes,inverse_flag=False):
    result = bedToolsFile.filter(featuretype_filter,featuretypes,inverse_flag=inverse_flag).saveas()
    return pybedtools.BedTool(result.fn)



```

```python
def parse_gtf_file(gtf_file):
    """
    given a gtf file as input this function will parse and return a single line
    is a generator : yield single parsed gtf entry one at a time
    """
    
    #sample_gtf_line (split in 3 lines for display purposes)
    #chromosome_1    Cufflinks       exon    20356   20687   .       +       .       
    #gene_id "XLOC_000036"; transcript_id "TCONS_00000050"; exon_number "1"; gene_name "g3"; oId 
    #"PAC:26903463"; contained_in "TCONS_00000051"; nearest_ref "PAC:26903463"; class_code "="; tss_id "TSS41"; p_id "P3";

    ##just creating an abtract class/container to hold the values for features of a gtf line
    class gtf_line_container(object):
        
        def __init__(self):
            self.chr=None
            self.start=None
            self.end=None
            self.transcript_id=None
            self.gene_id=None
            self.feature_type=None
            self.attributes=None
    
    for num,line in enumerate(open(gtf_file,'r')):
        
        line_split = line.strip().split('\t')
        
        #instantiate the container
        gtf_line = gtf_line_container()
        
        attributes_split=line_split[8].split(';')
        #sample attribute split line
        #['gene_id "XLOC_000001"', ' transcript_id "TCONS_00000001"', ' exon_number "1"', ' oId "CUFF.31.1"', ' class_code "u"', ' tss_id "TSS1"', '']
        
        #gene_id
        gene_id=attributes_split[0].split(' ')[1].replace('"','')
        
        #transcrpit_id
        transcript_id=attributes_split[1].split(' ')[2].replace('"','')
        
        
        gtf_line.chr=line_split[0].strip()
        gtf_line.start=line_split[3].strip()
        gtf_line.end=line_split[4].strip()
        gtf_line.strand=line_split[6].strip()
        gtf_line.feature_type=line_split[2].strip()
        gtf_line.gene_id = gene_id
        gtf_line.transcript_id = transcript_id
        gtf_line.attributes = line_split[8].strip()
        yield gtf_line





def summarize_cuffmerge_output(merged_gtf_file,summarize_by,outFile=None):
    '''
    given a merged.gtf file produced by cuffmerge this method can
    summarize it by gene or transcript level
    **PS: assumption is that for each gene the exons are reported, sorted by position**
    variable : summarize_by can take the value of gene,transcript
    '''

    features = {}
    class feature(object):
        def __init__(self):
            self.gene_id = None
            self.transcript_id = None
            self.start = None
            self.end = None
            self.strand = None
            self.exon_starts = []
            self.exon_ends = []
            self.attributes = None
            
    #default for each feature
    last_feature_id = None
    feature_start = 0;
    feature_chromosome = 'NA';
    feature_end = 0;
    feature_strand = 0;
    feature_attributes='NA'
    
    
    #setting the right feture name
    if summarize_by == 'gene':
        feature_name = 'gene_id'
    elif summarize_by == 'transcript':
        feature_name = 'transcript_id'
    else: 
        sys.stderr('Error : not a recognized summarize_by id %s ' % summarize_by)
        sys.exit(2)
    
    
    for gtf_line in parse_gtf_file(merged_gtf_file):
            #interesting peice of logic
            #depending on the user entered summarize_by varibale either the transcript_id or gene_id is used
            this_feature_id=eval('gtf_line.%s' % feature_name)
        
            #initialization
            if ( last_feature_id is None):
                last_feature_id = this_feature_id
                feature_start = gtf_line.start
                feature_end   = gtf_line.end
                feature_strand = gtf_line.strand
                feature_chromosome = gtf_line.chr
                feature_attributes = gtf_line.attributes
                continue
            
            #if we see the same feature, just update the end
            if ( last_feature_id == this_feature_id):
                feature_end   = gtf_line.end
                feature_attributes = gtf_line.attributes
                
            else:
                #create the new feature summary object before updating to the new feature
                feature_summary = feature()
                feature_summary.feature_id = last_feature_id
                feature_summary.start = feature_start
                feature_summary.end = feature_end
                feature_summary.strand = feature_strand
                feature_summary.chr = feature_chromosome
                feature_summary.attributes = feature_attributes
                
                #store the feature summary object in a dict
                features[last_feature_id] = feature_summary
                
                ##now update and store the new feature
                last_feature_id = this_feature_id
                feature_start = gtf_line.start
                feature_end   = gtf_line.end
                feature_strand = gtf_line.strand
                feature_chromosome = gtf_line.chr
                feature_attributes = gtf_line.attributes
    
    
    feature_summary = feature()
    feature_summary.feature_id = last_feature_id
    feature_summary.start = feature_start
    feature_summary.end = feature_end
    feature_summary.strand = feature_strand
    feature_summary.chr = feature_chromosome
    feature_summary.attributes = feature_attributes
                
    #store the feature summary object in a dict
    features[last_feature_id] = feature_summary
    
    #write the summarized file if needed
    if outFile is not None:
        fout = open(outFile,'w')
        for each_feature_id in features:
            feature = features[each_feature_id]
            gtf_line ='%s\t%s\t%s\t%s\t%s\t%s\t%s\t%s\t%s\n' %  (feature.chr,'cufflinks',summarize_by,feature.start,feature.end,
                                                                               '.',feature.strand,'.',feature.attributes)
            fout.write(gtf_line)
        print 'Created file %s summarized by %s of len %d' % (outFile,summarize_by,len(features))
        fout.close()
        return (features)
        
    print 'Created features dict, summarized by %s of len %d' % (summarize_by,len(features))
    return (features)
 
```

```python
#RNA-Seq genes
temp_rna_seq_genes_fh = tempfile.NamedTemporaryFile(delete=False)
rna_seq_genes = summarize_cuffmerge_output(rna_seq_assembly_file,summarize_by='gene',outFile=temp_rna_seq_genes_fh.name)
rna_seq_genes = pybedtools.BedTool(temp_rna_seq_genes_fh.name)

#RNA-Seq Transcripts
temp_rna_seq_transcripts_fh = tempfile.NamedTemporaryFile(delete=False)
rna_seq_transcripts = summarize_cuffmerge_output(rna_seq_assembly_file,summarize_by='transcript',outFile=temp_rna_seq_transcripts_fh.name)
rna_seq_transcripts = pybedtools.BedTool(temp_rna_seq_transcripts_fh.name)
```

```python
#available ref trasncriptome
cleaned_ref_transcriptome = pybedtools.BedTool(chlamy_ref_transcriptome).remove_invalid().saveas()

#cufflinks completed RNA-Seq assembly
rna_seq_assembly = pybedtools.BedTool(rna_seq_assembly_file)
```

```python
#number of novel genes predicted by RNA-Seq assembly
rna_seq_novel_genes = rna_seq_genes.intersect(ref_genes, v=True).saveas()

#number of novel antisense genes
rna_seq_antisense_novel_genes = rna_seq_genes.intersect(ref_genes, S=True).saveas()

#RNA-Seq novel exons
rna_seq_novel_exons = rna_seq_exons.intersect(ref_exons,v=True).saveas()

#RNA-Seq novel exons inside ref genes
rna_seq_novel_exons_inside_ref_genes = rna_seq_novel_exons.intersect(ref_genes).saveas()
```
