```python
#the following code reads the Chlamy v5.3 ref gene models
# and creates a dict mapping the chalmy common gene names to standard genenames
# this is needed to get the standard gene name from the cufflinks produced merged.gtf file which for some genes include the commonn gene name

def get_chlamy_common_genesName_to_standard_geneName_dict():
    chlamy_common_geneName_to_std_geneName = {}
    chlamy_v5_3_ref = '/global/projectb/projectdirs/PI/Chlamy_Chia_Lin/annotation/Chlamy_v5/Creinhardtii_v5.3_223_gene.gff3'
    for count,feature in enumerate(pybedtools.BedTool(chlamy_v5_3_ref)):
        if count == 100:
            pass
        #print feature[2]
        if feature[2] == 'mRNA':
            common_geneName = feature.attrs.get('geneName','NA')
            standard_transcriptName = feature.attrs.get('Name','NA')
            #removing the transcript prefix
            m = re.search('(^.+?).t.*$',standard_transcriptName)
            if m:
                standard_geneName = m.group(1)
            #chlamy_common_geneName_to_std_geneNa
            chlamy_common_geneName_to_std_geneName[common_geneName] = standard_geneName
    
    return chlamy_common_geneName_to_std_geneName
            
        
    
chlamy_common_geneName_to_std_geneName = get_chlamy_common_genesName_to_standard_geneName_dict()
```

```python
#a function to pass only features of a particular featuretype
#this is similar to grep
# if inverse_flag is set to True, this is similar to grep -v
def featuretype_filter(feature,featuretypes,inverse_flag=False):
    for featuretype in featuretypes:
        if feature[2] == featuretype:
            #match found
            if inverse_flag is True:
                return False
            else:
                return True
            
    #no match found
    if inverse_flag is True:
        return True
    else:
        return False 


#function to subset a feature file selecting the features 
def subset_featuretypes(bedToolsFile,featuretypes,inverse_flag=False):
    result = bedToolsFile.filter(featuretype_filter,featuretypes,inverse_flag=inverse_flag).saveas()
    return pybedtools.BedTool(result.fn)
```

```python
def create_pandas_df_from_gtf_file(merged_gtf_file):
    '''
    given a merged.gtf file produced by cuffmerge this method can
    summarize it by gene or transcript level
    **PS: assumption is that for each gene the exons are reported, sorted by position**
    variable : summarize_by can take the value of gene,transcript
    '''
    
    features = []
    for count,line in enumerate(pybedtools.BedTool(merged_gtf_file)):
        #parse following from the attrs
        attributes = ['gene_id','transcript_id','exon_number','gene_name','class_code','tss_id','p_id']
        attr_value = [ line.attrs.get(attr,'NA').strip() for attr in attributes ]
        feature = [ x.strip() for x in line[0:8]] + attr_value
        features.append(feature)
    
    #creating pandas df
    colnames = ['chr','source','feature','start','stop','score','strand','frame'] + attributes
    out_pandas_df = pandas.DataFrame.from_records(features,columns=colnames)
    
    return out_pandas_df
```

```python
def create_gtf_entry_from_df(group):
    output = []
    for index,row in group.T.iteritems():
        gtf_line = [row['chr'], row['source'], row['feature'], 
                    str(row['start']), str(row['stop']), row['score'],
                    row['strand'], str(row['frame'])]
        
        gtf_line_attrs = ['gene_id','transcript_id','exon_number','gene_name','tss_id','p_id']
        attr_line = ''
        
        for attr in gtf_line_attrs:
            value=str(row.get(attr,'NA'))
            if value.strip() != 'NA':
                attr_line += ('%s "%s"; ' % (attr,value))
        
        gtf_line.append(attr_line)
        output.append(gtf_line)
        
    return (output)
```
