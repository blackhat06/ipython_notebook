```python
def remove_low_coverage_reads(bamFile,coverage=2):        SUB ='remove_low_coverage_reads'        base_path = os.path.dirname(bamFile)    bamFile_name = os.path.basename(bamFile)    bamFile_handle = pysam.Samfile(bamFile,'rb')    references = bamFile_handle.references    out_bamFile = base_path + '/' + bamFile_name.replace('.bam','') + '_mincov_'+str(coverage)+'.bam'            if os.path.exists(out_bamFile):        print 'Expected output %s already present \n no processed required' % (out_bamFile)        return out_bamFile            out_bamFile_handle = pysam.Samfile(out_bamFile,'wb',template=bamFile_handle)    count_num_reads_low_coverage = 0;            for chromosome in references:        low_coverage_read_headers = {}                for pileupcol in bamFile_handle.pileup(chromosome):           if pileupcol.n <= coverage and pileupcol.n > 0:                for read in pileupcol.pileups:                    low_coverage_read_headers[read.alignment.qname] = 1                #if len(low_coverage_read_headers.keys()) > 0:            #print low_coverage_read_headers.keys()            #print                 for read in bamFile_handle.fetch(chromosome):            if ( low_coverage_read_headers.get(read.qname,'None') == 'None'):                out_bamFile_handle.write(read)            else:                count_num_reads_low_coverage += 1                os.system("samtools index out_bamFile")    print 'Wrote file %s, found and skipped %d low coverage reads' % (os.path.basename(out_bamFile), count_num_reads_low_coverage)        #doesn't work inside the     #index_bam(out_bamFile)        return out_bamFile
```
